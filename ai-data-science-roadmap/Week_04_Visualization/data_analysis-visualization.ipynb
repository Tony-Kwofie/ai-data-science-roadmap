```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your Titanic dataset
df = pd.read_csv(r"C:\Users\HP\Desktop\tony python\data set\titanic.csv")
```

```python
# Step 2: Basic Plot - Survivors Count
sns.countplot(x="Survived", data=df)
plt.title("Survivors Count (0 = No, 1 = Yes)")
plt.show()
```

```python
# Step 3: Bar Plot - Gender vs Survival
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Count by Gender")
plt.show()
```

```python
# Step 4: Scatter Plot - Age vs Fare

plt.figure(figsize=(8,6))
sns.scatterplot(x="Age", y="Fare", data=df)
plt.title("Age vs Fare Paid")
plt.show()

```

```python
# Step 5: Box Plot - Fare by Passenger Class
sns.boxplot(x="Pclass", y="Fare", data=df)
plt.title("Fare Distribution by Passenger Class")
plt.show()
```

```python
# Step 6: Heatmap - Missing Values
plt.figure(figsize=(10,6))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Data Heatmap")
plt.show()
```

```python
# Age Distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()

```

```python
# Average Age of Surviving vs Non-Surviving Passengers, by Gender
```

```python
# Step 1: Group and Prepare the Data

grouped_data = df.groupby(['Survived', 'Sex'])['Age'].mean().reset_index()
print(grouped_data)
```

```python
# Step 2: Create a Bar Plot

sns.barplot(x='Sex', y='Age', hue='Survived', data=grouped_data)
plt.title('Average Age by Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Average Age')
plt.legend(title='Survived (0=No, 1=Yes)')
plt.show()
```

```python
# Try a Pie Chart
# Let’s show the percentage of survivors vs non-survivors:

survival_counts = df['Survived'].value_counts()
labels = ['Did Not Survive', 'Survived']
colors = ['#FF9999','#66B3FF']

plt.pie(survival_counts, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Survival Rate - Titanic')
plt.axis('equal')  # Makes it a perfect circle
plt.show()
```
